## 1.8 Lets explore the variables

1. employee_id: Unique ID for employee
2. department: Department of employee and there are total of 9 departments
3. region:  Region of employment (unordered) and there are 34 regions
4. education: : Education Level
5. gender:  Gender of Employee
6. recruitment_channel: Channel of recruitment for employee
7. no_of_trainings: no of other trainings completed in previous year on soft skills, technical skills etc., here the minimum is 1 which means every employee is completing atleast one training 
8. age: Age of Employee, ranges from 20 to 60
9. previous_year_rating: Employee Rating for the previous year
10. length_of_service:  Length of service in years
11. KPIs_met >80%:  if Percent of KPIs(Key performance Indicators) >80% then 1 else 0
12. awards_won?: if awards won during previous year then 1 else 0
13. avg_training_score: Average score in current training evaluations
14. is_promoted: (Target) Recommended for promotion and binary

# 2. Data Analysis

## 2.1 How is the distribution of target labels? - Did most people get promoted
    
    Here, 1 is likely to get promoted and 0 is less likely to get promoted

```python
sns.countplot(train['is_promoted'])
```

### Well this is clearly Imbalanced
    Based on the description most of the people don't get promoted

## 2.2 Male and female employees ratio

```python
sns.countplot(train['gender'])
```

## 2.3 Who is get promoted most? Male or female?

```python
plt.figure(figsize=(11,9))
sns.countplot(x='gender', hue='is_promoted',data = train, palette="Set1")
```

### Males have relatively more difficulties in geting promoted, this could be because of the general larger population of male employees opposed to females employees.

## 2.4 Who has highest qualification education? M or F?

```python
plt.figure(figsize=(11,9))
sns.countplot(x='gender', hue='education',data = train, palette="Set1")
```

#### compared to female there are more number of master's and and above male employees 

## 2.5 In which department does most people get promoted?

```python
plt.figure(figsize=(11,9))
sns.countplot(x='department', hue='is_promoted',data = train, palette="Set1")
```

## 2.6 does length of service has any influence factor on promotion?

```python
sns.lmplot(x= 'age', y= 'length_of_service', hue = 'is_promoted', data = train)
```

## 2.7 Who is staying in the company M or F?


```python
plt.figure(figsize=(11,9))
sns.countplot(x='length_of_service', hue='gender',data = train, palette="Set1")
```

### Male employees are staying long with the company compared to the female employees

## 2.8 Does KPI score has influence on promotion

```python
plt.figure(figsize=(11,9))
sns.countplot(x='KPIs_met >80%', hue='is_promoted',data = train, palette="Set1")
```

## 2.9 Does awards has any influence on promotion

```python
plt.figure(figsize=(11,9))
sns.countplot(x='awards_won?', hue='is_promoted',data = train, palette="Set1")
```

```python
pd.crosstab(train['awards_won?'], train['is_promoted'], normalize='index')
```

## There is better chance to get promoted if award is won 

## 2.10 Does previous year rating infuence promotion?

```python
plt.figure(figsize=(11,9))
sns.countplot(x='previous_year_rating', hue='is_promoted',data = train, palette="Set1")
```

## 2.11 Based on education who has good avg training score and chance to get promoted?

```python
sns.relplot(x='avg_training_score',y= 'age', hue='is_promoted',col='education',data=train)
```

### Here clearly employees with training score of 90 or above, has a very good chance to get promoted 

## 2.12 relation ship between score and lenght of service to get promoted

```python
sns.relplot(x='avg_training_score',y= 'length_of_service', hue='is_promoted', col='education', data=train)
```

## 2.13 Relationship between age and length of service

```python
sns.relplot(x='age',y= 'length_of_service', hue='is_promoted', col='education',data=train)
```
